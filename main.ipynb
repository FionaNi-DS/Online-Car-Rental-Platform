```python
def main():
    shop = CarRental(10)   # initial stock = 10 cars
    customer = Customer()

    while True:
        print("\n===== Car Rental Menu =====")
        print("1. Display available cars")
        print("2. Rent cars on hourly basis ($5/hour)")
        print("3. Rent cars on daily basis ($20/day)")
        print("4. Rent cars on weekly basis ($60/week)")
        print("5. Return cars")
        print("6. Exit")

        choice = input("Enter your choice (1–6): ")

        if choice == "1":
            shop.display_stock()
        elif choice == "2":
            n = int(input("How many cars would you like to rent? "))
            customer.request_car("hourly", n, shop)
        elif choice == "3":
            n = int(input("How many cars would you like to rent? "))
            customer.request_car("daily", n, shop)
        elif choice == "4":
            n = int(input("How many cars would you like to rent? "))
            customer.request_car("weekly", n, shop)
        elif choice == "5":
            customer.return_car(shop)
        elif choice == "6":
            print("Thank you for using our service!")
            break
        else:
            print("Invalid choice, please try again.")
```

```python
main()
```

```python
# ==== Demo: stock -> rent -> (simulate time) -> return -> bill -> stock back ====
import importlib, car_rental
from datetime import timedelta

# reload to ensure latest code is used
importlib.reload(car_rental)
from car_rental import CarRental, Customer

print("\n===== DEMO START =====")

# 1) Create shop & customer, show initial stock
shop = CarRental(10)
cust = Customer()
print("\n[Step 1] Initial stock:")
shop.display_stock()   # Screenshot #1

# 2) Rent cars (choose a basis & quantity)
print("\n[Step 2] Customer rents 2 cars on a DAILY basis:")
cust.request_car("daily", 2, shop)  # prints confirmation
print("\nStock after renting (should be reduced):")
shop.display_stock()   # Screenshot #2 (reduced stock)

# 3) Simulate elapsed time so the bill is > 0 (here: 1 day 4 hours -> ceil to 2 days)
print("\n[Step 3] Simulate elapsed time (advance rental_time by 1 day 4 hours)")
cust.rental_info["rental_time"] -= timedelta(days=1, hours=4)

# 4) Return cars & show bill
print("\n[Step 4] Return cars & bill:")
bill = cust.return_car(shop)        # prints bill line inside return_cars
print("Final bill (returned value):", bill)  # optional echo
# Screenshot #3 (bill output)

# 5) Stock should be restored to initial amount
print("\n[Step 5] Stock after return (should be restored):")
shop.display_stock()   # Screenshot #4 (restored stock)

print("\n===== DEMO END =====")
```
